Cereal Diistribution

```python
plt.figure(figsize = (20,8))
plt.title("cereal data distribution")
sns.histplot(data = df)
plt.show()
```

```python
# creating a function which plot histogram chart for all the data 
def chart(c_df,c_name):
    plt.figure(figsize=(20,3))
    plt.title(f"Distribution of {c_name}")
    sns.histplot(c_df,kde=True)
    plt.xticks(rotation= 90)
    
```

```python
# creating a variable which stores all the column names
column_names = df.columns
```

```python
column_names
```

```python
for i in range(12):
    chart(df[column_names[i]],column_names[i])
```

```python
# ploting a graph which shows all the ratings of the cereals 
plt.figure(figsize = (20,5))
plt.title("Rating of the cerial")
sns.barplot(data = df,x='name',y='rating')
plt.xticks(rotation = 90)
plt.show()
```

```python
# creating a variable which stores the correlation values of the data
cor = df.corr()
```

```python
# ploting a heatmap with correlated values
plt.figure(figsize = (10,6))
sns.heatmap(cor,annot= True, cmap='coolwarm',fmt=".2f")
plt.show()
```
